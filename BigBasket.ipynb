## Find out the outliers from the dataset according to the columns and fill them with the mean.

### Removal of outliers according to columns

#### sale_price

```python
# Calculate iqr and outliers fence
q1=df["sale_price"].quantile(0.25)
q3=df["sale_price"].quantile(0.75)
iqr=q3-q1
iqr
```

```python
lower_fence=max(q1-1.5*iqr,0)
upper_fence=q3+1.5*iqr

lower_fence
```

```python
upper_fence
```

```python
# To detect the number of outliers present in sale price
outliers=df[(df["sale_price"]< lower_fence) | (df["sale_price"] >
upper_fence)]
print("Outliers_count:", len(outliers) )
```

```python
# removing the outliers
df['sale_price' ]=np.where(df['sale_price'] <= lower_fence,lower_fence,df[
'sale_price' ])
df['sale_price']=np.where(df['sale_price']>=upper_fence, upper_fence,df[
'sale_price' ])

#Boxplot after caping of outliers
sns.boxplot(df['sale_price' ], color="red")
plt.title("After caping of outliers")
plt.show()
```

```python
#checking of outliers after caping in sale_price
outliers_after_caping = df[(df["sale_price"] < lower_fence) &
(df["sale_price"] > upper_fence)]
print("outliers left after caping in sale_price:", outliers_after_caping.shape[0])
```

#### market_price

```python
# Calculate outliers and outer fence
q1=df["market_price"].quantile(0.25)
q3=df["market_price"].quantile(0.75)
iqr=q3-q1
iqr
```

```python
lower_fence=max((q1-1.5*iqr,0))
upper_fence=q3+1.5*iqr
lower_fence
```

```python
upper_fence
```

```python
# To detect the number of outliers present in market_price
outliers=df[(df["market_price"]< lower_fence) | (df["market_price"] >
upper_fence) ]
print("Outliers_count:", len(outliers))
```

```python
# Caping of outliers
df['market_price' ]=np.where(df['market_price'] <= lower_fence, lower_fence,df['market_price'])
df['market_price' ]=np.where(df['market_price' ]>=upper_fence, upper_fence,df['market_price' ])

# Showing boxplot after caping of outliers
sns.boxplot(df["market_price"], color="yellow")
plt.title("After caping outliers")
plt.show()
```

```python
#checking of outliers after caping in market price
outliers_after_caping = df[(df["market_price"] < lower_fence) &
(df["market_price"] > upper_fence)]
print("outliers left after caping in market_price:", outliers_after_caping.shape[0])
```

#### Rating

```python
# Calculate outliers and outer fence of rating
q1=df["rating"].quantile(0.25)
q3=df["rating"].quantile(0.75)
iqr=q3-q1
iqr
```

```python
lower_fence=q1-1.5*iqr
upper_fence=q3+1.5*iqr
lower_fence
```

```python
upper_fence
```

```python
# Caping of outliers
df["rating"]=np.where(df["rating"]<lower_fence, lower_fence,df["rating"]
)
df["rating"]=np.where(df["rating"]>upper_fence, upper_fence,df["rating"]
)

# Boxplot for rating after caping outliers
sns.boxplot(df["rating"], color="pink")
plt.title("After caping outliers")
plt.show()
```

```python
#checking of outliers after caping in rating
outliers_after_caping = df[(df["rating"] < lower_fence) & (df["rating"]> upper_fence)]
print("outliers left after caping in rating:", outliers_after_caping.shape[0])
```
