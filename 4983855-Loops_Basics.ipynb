## Monthly Bill Tracker

```python
# List of monthly bills
bills = [1500, 2000, 3000]
bills = map(int,input().split())

# Initialize total
total = 0

# Iterate through the list
for bill in bills:
    print("Bill amount:", bill)
    total += bill

# Print total amount
print("Total of monthly bills:", total)
```

## School Attendance

##  Salary Distribution

```python
# Dictionary of employees and their bonuses
bonuses = {"John": 5000, "Mary": 6000}

# Iterate through the dictionary
for name, bonus in bonuses.items():
    print(f"Name: {name} - Bonus: ₹{bonus}")
```

## . Grade Report

```python
# List of student scores
scores = [85, 92, 76, 61]

# Calculate and print grades
for i, score in enumerate(scores, start=1):
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    else:
        grade = "F"
    
    print(f"Student {i}: Score = {score}, Grade = {grade}")
```

## Book Availability

```python
# List of books
books = ["Python Basics", "Data Science"]

# Ask user for book title
title = input("Enter the book title: ")

# Check availability
if title in books:
    print("Available")
else:
    print("Not Available")
```

## Temperature Analysis

```python
# List of daily temperatures
temperatures = [30, 32, 28, 35, 29]

# Assume first value as highest and lowest
highest = temperatures[0]
lowest = temperatures[0]

# Loop through the list
for temp in temperatures:
    if temp > highest:
        highest = temp
    if temp < lowest:
        lowest = temp

# Print results
print("Highest Temperature:", highest)
print("Lowest Temperature:", lowest)
```

## Inventory Management


```python
# Dictionary of items and their quantities
stock = {"Apples": 5, "Bananas": 20}

# Check for low-stock items (quantity below 10)
for item, quantity in stock.items():
    if quantity < 10:
        print(f"{item} is low in stock (Quantity: {quantity})")
```

## Flight Seat Booking


```python
# List of all seats
seats = [1, 2, 3, 4]

# List of booked seats
booked_seats = [2, 4]

# Check each seat
for seat in seats:
    if seat in booked_seats:
        print(f"Seat {seat}: Booked")
    else:
        print(f"Seat {seat}: Available")
```
