```python
temperature = float(input("Enter patient's temperature in °F: "))
pain_level = input("Enter patient's pain level (none, moderate, severe): ").lower()

if temperature > 102 or pain_level == "severe":
    priority = "High Priority"
elif 99 <= temperature <= 102 or pain_level == "moderate":
    priority = "Medium Priority"
else:
    priority = "Low Priority"

print("Patient Priority:", priority)
```

```python
cart_value = float(input("Enter total cart value (₹): "))
membership = input("Do you have a membership? (yes/no): ").lower()

if cart_value > 2000:
    Delivery = "Free Delivery"
elif 1000 <= cart_value <= 2000 and membership == "yes":
    Delivery = "Free Delivery"
else:
    Delivery = "Delivery charge is ₹100"

print(Delivery)
```

```python
overdue_days = int(input("Enter number of overdue days: "))

if overdue_days <= 5:
    fine = overdue_days * 2
elif 6 <= overdue_days <= 10:
    fine = overdue_days * 5
else:
    fine = overdue_days * 10

print("Library Fine: ₹", fine)
```

```python
class_type = input("Enter class type (economy, business, first): ").lower()
luggage_weight = float(input("Enter luggage weight: "))

excess_fee = 0

if class_type == "economy":
    if luggage_weight > 20:
        excess_fee = (luggage_weight - 20) * 200
elif class_type == "business":
    if luggage_weight > 30:
        excess_fee = (luggage_weight - 30) * 150
elif class_type == "first":
    if luggage_weight > 40:
        excess_fee = (luggage_weight - 40) * 100
else:
    print("Invalid class type entered.")


if class_type in ["economy", "business", "first"]:
    if excess_fee > 0:
        print("Excess Luggage Fee: ₹", excess_fee)
    else:
        print("No excess luggage fee. You're within the limit.")
```

```python
customer_type = input("Enter customer type (regular, member, VIP): ").lower()
room_type = input("Enter room type (standard, deluxe): ").lower()
room_price = float(input("Enter room price (₹): "))

discount = 0

if customer_type == "regular":
    discount = 0.10 * room_price
elif customer_type == "member":
    if room_type == "deluxe":
        discount = 0.20 * room_price
    elif room_type == "standard":
        discount = 0.15 * room_price
    else:
        print("Invalid room type entered.")
elif customer_type == "vip":
    discount = 0.30 * room_price
else:
    print("Invalid customer type entered.")


final_price = room_price - discount


print("Final price after discount: ₹", final_price)
```

```python
units = int(input("Enter total Consumption: "))


bill = 0


if units <= 100:
    bill = units * 5
elif units <= 300:
    # First 100 units at ₹5, remaining at ₹10
    bill = (100 * 5) + ((units - 100) * 10)
else:
    # First 100 units at ₹5, next 200 units at ₹10, remaining at ₹15
    bill = (100 * 5) + (200 * 10) + ((units - 300) * 15)


print("Total Electricity Bill: ₹", bill)
```

```python
time_of_day = input("Enter time of day (day/night): ").lower()
distance = float(input("Enter distance traveled in km: "))


if time_of_day == "day":
    rate_per_km = 15
elif time_of_day == "night":
    rate_per_km = 20
else:
    print("Invalid time of day entered.")
    rate_per_km = 0


if rate_per_km > 0:
    fare = distance * rate_per_km
    

    if distance > 20:
        fare *= 0.90  # 10% discount
  
    print("Total Fare: ₹", fare)
```

```python
age = int(input("Enter Your Age: "))
student_id = input("Do you have a student ID? (yes/no): ").lower()
full_price = float(input("Enter full entry price (₹): "))


entry_fee = full_price


if age < 12 or age > 60:
    entry_fee = 0
elif 12 <= age <= 25 and student_id == "yes":
    entry_fee = full_price * 0.5
else:
    entry_fee = full_price


if entry_fee == 0:
    print("Entry is free!")
else:
    print("Entry Fee: ₹", entry_fee)
```

```python
vehicle_type = input("Enter vehicle type (car/bike): ").lower()
hours = float(input("Enter parking duration in hours: "))

parking_fee = 0


if vehicle_type == "car":
    if hours <= 1:
        parking_fee = 50
    else:
        parking_fee = 50 + (hours - 1) * 20
elif vehicle_type == "bike":
    if hours <= 1:
        parking_fee = 20
    else:
        parking_fee = 20 + (hours - 1) * 10
else:
    print("Invalid vehicle type entered.")


if vehicle_type in ["car", "bike"]:
    print("Parking Fee: ₹", parking_fee)
```

```python
monthly_income = float(input("Enter monthly income (in Rupees): "))
existing_loan = input("Do you have existing loans? (yes/no): ").lower()


ltv_ratio = 0


if monthly_income > 100000 and existing_loan == "no":
    ltv_ratio = 90
elif 50000 <= monthly_income <= 100000 or existing_loan == "yes":
    ltv_ratio = 70
else:
    ltv_ratio = 50


print("Eligible Loan-to-Value (LTV) Ratio:", ltv_ratio, "%")
```
