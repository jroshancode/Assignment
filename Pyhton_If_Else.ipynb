## Discount Calculator


```python
# Take input from user
price = float(input("Enter the price of the product: ₹"))

# Apply discount
if price > 500:
    discount = price * 0.10   # 10% discount
else:
    discount = price * 0.05   # 5% discount

# Calculate final price
final_price = price - discount

# Display result
print("Discount applied: ₹", discount)
print("Final price after discount: ₹", final_price)
```

## Evening Discount

```python
# Take product price and current time as input
price = float(input("Enter the product price: ₹"))
time = int(input("Enter current hour (24-hour format, 0–23): "))

# Check if time is between 6 PM (18) and 9 PM (21)
if 18 <= time <= 21:
    discount = price * 0.20   # 20% discount
else:
    discount = 0             # No discount

# Calculate final price
final_price = price - discount

# Display result
print("Discount applied: ₹", discount)
print("Final price: ₹", final_price)
```

## Restaurant Bill Split


```python
# Take inputs
people = int(input("Enter number of people in the group: "))
bill = float(input("Enter total bill amount: ₹"))

# Check if service charge applies
if people > 5:
    service_charge = bill * 0.15
    final_amount = bill + service_charge
    print("15% service charge applied.")
    print("Final bill amount: ₹", final_amount)
else:
    print("No service charge applied.")
    print("Final bill amount: ₹", bill)
```

## BMI Calculator

```python
# Take weight and height as input
weight = float(input("Enter your weight (in kg): "))
height = float(input("Enter your height (in meters): "))

# Calculate BMI
bmi = weight / (height ** 2)

# Display BMI
print("Your BMI is:", round(bmi, 2))

# Check BMI category
if bmi < 18.5:
    print("Underweight")
elif 18.5 <= bmi <= 24.9:
    print("Normal")
else:
    print("Overweight")
```

## Flight Fare Check

```python
# Take inputs
age = int(input("Enter passenger age: "))
fare = float(input("Enter ticket fare: ₹"))

# Check discount eligibility
if age < 12 or age > 60:
    discount = fare * 0.50
    final_fare = fare - discount
    print("50% discount applied.")
    print("Final fare: ₹", final_fare)
else:
    print("No discount applied.")
    print("Final fare: ₹", fare)
```

##  Electricity Bill

```python
# Take units consumed as input
units = int(input("Enter number of electricity units consumed: "))

# Calculate bill based on slabs
if units <= 100:
    bill = units * 5
elif units <= 300:
    bill = units * 10
else:
    bill = units * 15

# Display total bill
print("Total electricity bill: ₹", bill)
```
