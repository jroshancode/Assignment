```python
correct_pin = "9876"
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    pin = input("Enter your PIN: ")

    if pin == correct_pin:
        print("PIN accepted. Access granted.")
        break
    else:
        attempts += 1
        print("Incorrect PIN.")

if attempts == max_attempts:
    print("Card blocked. Too many incorrect attempts.")
```

```python
while True:
    password = input("Enter a password: ")

    # Check length
    if len(password) < 8:
        print("Password must be at least 8 characters long.")
        continue

    # Check for letters and numbers
    has_letter = False
    has_number = False

    for ch in password:
        if ch.isalpha():
            has_letter = True
        if ch.isdigit():
            has_number = True

    if has_letter and has_number:
        print("Valid password accepted.")
        break
    else:
        print("Password must contain both letters and numbers.")
```

```python
days = int(input("Enter number of days until plan expires: "))

while days > 0:
    print("Reminder: Your plan expires in", days, "day(s).")
    days -= 1

print("Plan Expired.")

```

```python
while True:
    email = input("Enter your email address: ")

    if "@" in email and "." in email:
        print("Valid email address")
        break
    else:
        print("Invalid email. Please try again.")
```

```python
loan_amount = float(input("Enter loan amount: "))
monthly_payment = float(input("Enter monthly payment: "))

while loan_amount > 0:
    loan_amount -= monthly_payment

    if loan_amount < 0:
        loan_amount = 0

    print("Remaining loan balance: ₹", loan_amount)

print("Loan fully repaid.")
```

```python
questions = [
    ("What is 2 + 3? ", 5),
    ("What is 10 - 4? ", 6),
    ("What is 6 * 2? ", 12),
    ("What is 20 / 5? ", 4),
    ("What is 9 + 1? ", 10)
]

i = 0

while i < len(questions):
    answer = int(input(questions[i][0]))

    if answer == questions[i][1]:6
        print("Correct! Correct Answer:",answer)
        i += 1
    else:
        print("Wrong answer. Game over!")
        break

if i == len(questions):
    print("Congratulations! You answered all 5 questions correctly.")
```

```python
temperature = -12      # initial temperature
threshold = -5         # maximum safe temperature

while temperature <= threshold:
    print("Current temperature:", temperature, "°C")
    temperature += 1   # simulate temperature rising each minute

print("WARNING: Temperature exceeded safe limit!")
print("Current temperature:", temperature, "°C")
```

```python
available_spaces = 10

while available_spaces > 0:
    print("Available parking spaces:", available_spaces)
    input("Car arrived. Press Enter to park the car.")
    available_spaces -= 1

print("Parking Full. No spaces left.")
```

```python
vehicle_count = 0

while True:
    entry = input("Vehicle passed? Type 'yes' or 'stop': ")

    if entry.lower() == "stop":
        break
    elif entry.lower() == "yes":
        vehicle_count += 1
    else:
        print("Invalid input. Please type 'yes' or 'stop'.")

print("Total number of vehicles passed:", vehicle_count)
```

```python
budget = float(input("Enter your total budget: "))

while budget > 0:
    price = float(input("Enter item price: "))

    if price <= budget:
        budget -= price
        print("Item purchased.")
        print("Remaining budget: ₹", budget)
    else:
        print("Not enough budget to buy this item.")

print("Budget exhausted. Shopping finished.")
```
