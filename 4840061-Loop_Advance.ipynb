```python
# Sample temperature data for 3 cities over 7 days
weather_data = {
    "City A": [30, 32, 28, 35, 29, 31, 33],
    "City B": [25, 27, 26, 30, 24, 28, 29],
    "City C": [20, 22, 19, 23, 21, 20, 24]
}

for city, temps in weather_data.items():
    highest = temps[0]
    lowest = temps[0]
    
    for temp in temps:  # Iterate over each day
        if temp > highest:
            highest = temp
        if temp < lowest:
            lowest = temp
            
    print(f"{city}: Highest Temperature = {highest}, Lowest Temperature = {lowest}")
```

```python
import string

# Sample list of passwords
passwords = ["Pass123!", "hello", "Wealth@2025", "weakpass", "Abc!"]

special_chars = string.punctuation  # All special characters

for pwd in passwords:
    length_ok = len(pwd) >= 8
    has_number = any(char.isdigit() for char in pwd)
    has_special = any(char in special_chars for char in pwd)
    
    print(f"Password: {pwd}")
    
    if length_ok and has_number and has_special:
        print(" - Strong password ✅")
    else:
        print(" - Weak password ❌")
        if not length_ok:
            print("   * Password should be at least 8 characters long")
        if not has_number:
            print("   * Password should contain at least one number")
        if not has_special:
            print("   * Password should contain at least one special character")
    print()
```

```python
# Initialize a 3x3 grid with empty spaces
grid = [[" " for _ in range(3)] for _ in range(3)]

def display_grid():
    for row in grid:
        print(" | ".join(row))
        print("-" * 9)

# Game loop for marking cells
while True:
    display_grid()
    
    # Ask the user for input
    row = int(input("Enter row number (1-3) or 0 to exit: ")) - 1
    if row == -1:
        print("Exiting the game.")
        break
    col = int(input("Enter column number (1-3): ")) - 1
    mark = input("Enter your mark (X or O): ").upper()
    
    # Validate input
    if row not in range(3) or col not in range(3):
        print("Invalid row or column. Try again.\n")
        continue
    if mark not in ["X", "O"]:
        print("Invalid mark. Use X or O.\n")
        continue
    if grid[row][col] != " ":
        print("Cell already taken. Choose another.\n")
        continue
    
    # Mark the cell
    grid[row][col] = mark
    print()
```

```python
# List of strings to check
strings = ["level", "hello", "level", "television", "madam"]

for s in strings:
    # Remove spaces and convert to lowercase for accurate checking
    clean_s = s.replace(" ", "").lower()
    
    if clean_s == clean_s[::-1]:
        print(f"'{s}' is a palindrome ✅")
    else:
        print(f"'{s}' is not a palindrome ❌")
```

```python
# List of employees
employees = [f"Employee {i}" for i in range(1, 11)]
days = 5

# Dictionary to store attendance
attendance = {employee: [] for employee in employees}

# Record attendance
for employee in employees:
    print(f"Mark attendance for {employee}:")
    for day in range(1, days + 1):
        while True:
            mark = input(f"  Day {day} (P/A): ").upper()
            if mark in ["P", "A"]:
                attendance[employee].append(mark)
                break
            else:
                print("Invalid input. Enter 'P' for present or 'A' for absent.")

# Display attendance summary
print("\nAttendance Summary:")
for employee, marks in attendance.items():
    present_count = marks.count("P")
    absent_count = marks.count("A")
    print(f"{employee}: Present = {present_count}, Absent = {absent_count}, Record = {marks}")
```

```python
products = ["Product A", "Product B", "Product C"]
months = ["Month 1", "Month 2", "Month 3", "Month 4"]


sales = []

print("Enter sales data:")
for i in range(len(products)):
    product_sales = []
    for j in range(len(months)):
        value = int(input(f"Enter sales of {products[i]} for {months[j]}: "))
        product_sales.append(value)
    sales.append(product_sales)

print("\nSales Data:")
print("Product   ", end="")
for m in months:
    print(m, end="   ")
print()

for i in range(len(products)):
    print(products[i], end="   ")
    for j in range(len(months)):
        print(sales[i][j], end="     ")
    print()

# Highlight highest sales each month
print("\nHighest Sales Each Month:")
for j in range(len(months)):
    max_sale = sales[0][j]
    max_product = products[0]

    for i in range(1, len(products)):
        if sales[i][j] > max_sale:
            max_sale = sales[i][j]
            max_product = products[i]

    print(f"{months[j]}: {max_product} with sales {max_sale}")
```

```python
# Read 3x3 matrix
matrix = []
print("Enter elements of 3x3 matrix:")

for i in range(3):
    row = []
    for j in range(3):
        row.append(int(input(f"Enter element [{i}][{j}]: ")))
    matrix.append(row)

#  It will show the Magic sum (sum of first row)
magic_sum = 0
for j in range(3):
    magic_sum += matrix[0][j]

is_magic = True

# Code for Checking rows
for i in range(3):
    row_sum = 0
    for j in range(3):
        row_sum += matrix[i][j]
    if row_sum != magic_sum:
        is_magic = False

# Checking columns
for j in range(3):
    col_sum = 0
    for i in range(3):
        col_sum += matrix[i][j]
    if col_sum != magic_sum:
        is_magic = False

# Check diagonals
diag1 = 0
diag2 = 0
for i in range(3):
    diag1 += matrix[i][i]
    diag2 += matrix[i][2 - i]

if diag1 != magic_sum or diag2 != magic_sum:
    is_magic = False

if is_magic:
    print("The matrix is a Magic Square ✅")
else:
    print("The matrix is NOT a Magic Square ❌")
```

```python
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
meals = ["Breakfast", "Lunch", "Dinner"]

# Initializing empty meal planner
planner = []

# Input meals
print("Enter your meals for the week:\n")
for day in days:
    day_meals = []
    print(f"{day}:")
    for meal in meals:
        item = input(f"  {meal}: ")
        day_meals.append(item)
    planner.append(day_meals)
    print()

# Display meal planner
print("\nWeekly Meal Planner:")
for i in range(len(days)):
    print(days[i])
    for j in range(len(meals)):
        print(f"  {meals[j]}: {planner[i][j]}")
    print()

# Modify meals
choice = input("Do you want to modify any meal? (yes/no): ")

while choice.lower() == "yes":
    day_index = int(input("Enter day number (1-7): ")) - 1
    meal_index = int(input("Enter meal number (1-Breakfast, 2-Lunch, 3-Dinner): ")) - 1

    new_meal = input("Enter new meal: ")
    planner[day_index][meal_index] = new_meal

    print("Meal updated successfully!\n")

    choice = input("Do you want to modify another meal? (yes/no): ")

# Final planner
print("\nFinal Meal Planner:")
for i in range(len(days)):
    print(days[i])
    for j in range(len(meals)):
        print(f"  {meals[j]}: {planner[i][j]}")
    print()
```
