## Hospital Emergency Room Triage


```python
# Take inputs
temperature = float(input("Enter patient's temperature (°F): "))
pain = input("Enter pain level (none/moderate/severe): ").strip().lower()

# Determine priority
if temperature > 102 or pain == "severe":
    print("High Priority")
elif (99 <= temperature <= 102) or pain == "moderate":
    print("Medium Priority")
else:
    print("Low Priority")

```

## Library Fine System

```python
# Take number of overdue days as input
days = int(input("Enter number of overdue days: "))

# Calculate fine based on days
if days <= 5:
    fine = days * 2
elif days <= 10:
    fine = days * 5
else:
    fine = days * 10

# Display the fine
print("Total fine: ₹", fine)
```

## Airport Luggage Check


```python
# Take inputs
class_type = input("Enter class type (economy/business/first): ").strip().lower()
weight = float(input("Enter luggage weight (in kg): "))

# Initialize limits and charges
if class_type == "economy":
    limit = 20
    charge_per_kg = 200
elif class_type == "business":
    limit = 30
    charge_per_kg = 150
elif class_type == "first":
    limit = 40
    charge_per_kg = 100
else:
    print("Invalid class type.")
    exit()

# Calculate excess fee
if weight > limit:
    excess_weight = weight - limit
    excess_fee = excess_weight * charge_per_kg
    print("Excess luggage fee: ₹", excess_fee)
else:
    print("No excess luggage fee.")
```

##  Hotel Room Pricing

```python
# Take inputs
customer_type = input("Enter customer type (regular/member/VIP): ").strip().lower()
room_type = input("Enter room type (standard/deluxe): ").strip().lower()
price = float(input("Enter room price: ₹"))

# Determine discount
if customer_type == "regular":
    discount = 0.10

elif customer_type == "member":
    if room_type == "deluxe":
        discount = 0.20
    elif room_type == "standard":
        discount = 0.15
    else:
        print("Invalid room type.")
        exit()

elif customer_type == "vip":
    discount = 0.30

else:
    print("Invalid customer type.")
    exit()

# Calculate final price
final_price = price - (price * discount)

# Display result
print("Final price after discount: ₹", final_price)
```

## Energy Consumption Billing

```python
# Take total consumption as input
units = int(input("Enter total electricity consumption (in units): "))

# Calculate bill using slab system
if units <= 100:
    bill = units * 5

elif units <= 300:
    bill = (100 * 5) + (units - 100) * 10

else:
    bill = (100 * 5) + (200 * 10) + (units - 300) * 15

# Display total bill
print("Total electricity bill: ₹", bill)
```

## Ride Fare Calculation

```python
# Take inputs
time_of_day = input("Enter time (day/night): ").strip().lower()
distance = float(input("Enter distance travelled (in km): "))

# Set rate based on time
if time_of_day == "day":
    rate = 15
elif time_of_day == "night":
    rate = 20
else:
    print("Invalid time entered.")
    exit()

# Calculate fare
fare = distance * rate

# Apply discount if distance > 20 km
if distance > 20:
    discount = fare * 0.10
    fare = fare - discount
    print("10% discount applied.")

# Display total fare
print("Total fare: ₹", fare)
```

## Event Entry Rules

```python
# Take inputs
age = int(input("Enter your age: "))
student_id = input("Do you have a valid student ID? (yes/no): ").strip().lower()
ticket_price = float(input("Enter ticket price: ₹"))

# Determine entry fee
if age < 12 or age > 60:
    final_price = 0
    print("Free entry.")

elif 12 <= age <= 25 and student_id == "yes":
    final_price = ticket_price * 0.50
    print("50% student discount applied.")

else:
    final_price = ticket_price
    print("Full price applicable.")

# Display final fee
print("Entry fee: ₹", final_price)
```

## Parking Lot Charges

```python
# Take inputs
vehicle = input("Enter vehicle type (car/bike): ").strip().lower()
hours = int(input("Enter parking duration (in hours): "))

# Calculate parking fee
if vehicle == "car":
    if hours >= 1:
        fee = 50 + (hours - 1) * 20
    else:
        fee = 50

elif vehicle == "bike":
    if hours >= 1:
        fee = 20 + (hours - 1) * 10
    else:
        fee = 20

else:
    print("Invalid vehicle type.")
    exit()

# Display total fee
print("Total parking fee: ₹", fee)
```

## Real Estate Loan Eligibility


```python
# Take inputs
income = float(input("Enter your monthly income (₹): "))
existing_loan = input("Do you have any existing loans? (yes/no): ").strip().lower()

# Determine LTV ratio
if income > 100000 and existing_loan == "no":
    ltv = 90

elif 50000 <= income <= 100000 or existing_loan == "yes":
    ltv = 70

else:
    ltv = 50

# Display eligible LTV ratio
print("Eligible Loan-to-Value (LTV) ratio:", ltv, "%")
```
