## Data Cleaning 
####  Check for missing values and handle them appropriately.
####  Remove duplicate entries if any.
####  Correct any inconsistencies or errors in the data.

```python
df.isna().sum()
```

### Filling the null value of duration column

```python
# To rectify the error of rating-duration
layer=df["rating"].str.contains("min" , na = False)

# Move duration values into duration column that were wrongly inputted in rating column
df.loc[layer , "duration"] = df.loc[layer, "rating"]

# Removing wrongly inputed rating values
df.loc[layer, "rating"] = None
```

### Analysis - Correction of Rating and Duration Values
. In this step, errors in the rating column were identified where duration values (such as minutes) were incorrectly recorded as ratings.

. A filter was applied to detect entries containing the term "min" within the rating column.

. These incorrect values were then transferred to the duration column to ensure accurate representation of content length. After relocation, the  erroneous values in the rating column were removed by setting them to null.

. This correction helps eliminate data duplication and improves the reliability of both the rating and duration attributes, ensuring cleaner and
   more accurate data for further analysis.

```python
df[df[["rating" , "duration"]].isna().any(axis=1)]
```

### Fill the real value which is Missing Rating in column
· In this step, the missing values in the rating column were addressed by manually verifying the correct ratings from reliable web and internet sources.

· Specific titles with null rating values were identified, and their accurate content ratings were researched and updated accordingly.

· This approach ensures that the rating column reflects correct classification standards, which is essential for meaningful content analysis an reliable insights in subsequent stages of the project.

## Correlation Analysis:
#### Investigate potential correlations between variables (e.g., ratings and duration)

```python
# Separate 'duration' by type
movies = df[df['type'] == 'Movie'].copy()
tv_shows = df[df['type'] == 'TV Show' ].copy()

# Extract duration in minutes (Movie)
movies['duration_minutes' ] =movies['duration' ].str.extract('(\d+)').astype(float)

# Extract number of seasons (TV Show)
tv_shows['seasons' ] = tv_shows['duration' ].str.extract('(\d+)').astype(float)
# Assign ordinal values to ratings (example scale)
rating_order = {'TV-Y': 1, 'TV-G': 2, 'G': 3, 'TV-PG': 4, 'PG': 5,
'PG-13': 6, 'TV-14': 7, 'R': 8, 'NC-17': 9, 'TV-MA': 10}

# Map ratings to numeric scale
movies['rating_score' ] = movies['rating' ].map(rating_order)
tv_shows['rating_score' ] = tv_shows['rating' ].map(rating_order)

# For Movies
plt.figure(figsize=(6, 4))
sns.heatmap(movies[['duration_minutes', 'rating_score' ]].corr(), annot=True,
cmap='YlGnBu')
plt.title('Correlation (Movies): Duration vs Rating')
plt.show()

# For TV Shows
plt.figure(figsize=(6, 4))
sns.heatmap(tv_shows[['seasons', 'rating_score' ]].corr(), annot=True,
cmap='YlOrRd' )
plt.title('Correlation (TV Shows): Seasons vs Rating')
plt.show()
```

## Content Evolution Over Time
#### Explore how the characteristics of content (e.g., duration, ratings) have evolved over the years.

```python
tv_shows = df[df["type"] == "TV Show"].copy()
tv_shows ["seasons" ] =tv_shows["duration"].str.extract(r'(\d+)').astype(float)

avg_seasons =tv_shows.groupby("release_year")["seasons"].mean().reset_index()

# Plot 1. for average TV show seasons over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_seasons, x="release_year", y="seasons",
marker="s", mfc="navy", mec="purple", color="coral", linewidth=2)
plt.title("Average TV Show Seasons Over the Years")
plt.xlabel("Release Year")
plt.ylabel("Avg Number of Seasons")
plt.tight_layout ()
plt.show()
```

```python
# Expand shorter seasonal series if they maintain high engagement and cost less.
# Renew long-season series only if viewership justifies it.
# Use this trend to balance binge-worthy short series with a few flagship multi-season originals.
rating_by_year = df.groupby(["release_year",
                               "rating"]).size().reset_index(name='count')

# Filter top 6 ratings for clarity
top_ratings = df["rating"].value_counts().head(6).index
filtered = rating_by_year[rating_by_year["rating"].isin(top_ratings)]

# Lineplot 2. for content rating trends over time
plt.figure(figsize=(14, 6))
sns.lineplot(data=filtered, x="release_year", y="count", hue="rating",
marker="o", linestyle="dotted")
plt.title("Content Ratings Trend Over Time")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.legend(title="Rating")
plt.tight_layout ()
plt. show()
```

## Finding,Conclusion and Recommendations

### KEY FINDINGS FROM NETFLIX DATASET:

Content Type & Duration: Movies dominate the catalog.Average movie duration has
remained mostly steady around 90-100 minutes, with slight variations across years.

    
Content Trends Over Time: A surge in new titles around 2018-2019, peaking just
before the COVID-19 pandemic. A dip in 2020-2021, likely due to pandemic-related
production halts.

Correlation Analysis Weak correlation between content duration and rating. No strong
relationship suggesting that longer movies or more seasons lead to stricter or looser
ratings.

Content Type & Duration: Netflix focuses on snackable movie content and limited-
series TV shows, likely due to budget, engagement, and bingeability factors.

Country & Language Distribution: United States is the top producer, followed by India,
UK, and Canada.English is the most used language, with Hindi, Spanish, French, and
Japanese as notable regional languages.

Genre Popularity: Top genres: International Movies, Dramas, Comedies, and
Documentaries.Genres like Children & Family Movies and Romantic Movies are also widely
represented.

Time-Based Content Trends: Netflix scales content supply around user availability and
holiday seasons.

### CONCLUSION:
From the insights and key findings we can conclude that Netflix offers a diverse range of content, with a strong focus on international and drama-based titles.The content library is heavily English-dominated, but localization efforts are evident, especially in India and Latin America. Content creation and release peaked in 2019, then declined temporarily during the pandemic.Netflix maintains a balanced content length, making content more accessible
for various user preferences.

### RECOMMENDATIONS:
1. Localization & Language Expansion: Increase availability of subtitles and dubbing in
regional languages.Promote multi-language content to broaden user base and accessibility.

2. Content Strategy Focus on strengthening regional content (e.g., Hindi, Korean,
Spanish) for deeper market penetration.Continue investing in documentaries and family
content, as they have consistent appeal.

3. Genre Personalization: Use data to personalize recommendations by genre and
duration preferences (e.g., short comedies vs. long dramas).

4. Seasonal Releases & Timing: Capitalize on seasonal trends (e.g., family content during
holidays, thrillers in fall).Encourage short-form content for mobile-first markets.


5. Monitoring Trends: Track user engagement over time to understand evolving
preferences (e.g., rising popularity of tech-themed genres, true crime).
