```python
bills = [1500, 2000, 3000]
total = 0

for b in bills:
    print("Bill amount:", b)
    total += b

print("Total monthly bills:", total)
```

```python
scores = [85, 92, 76, 61]

for score in scores:
    if score >= 90:
        print(score, "- Grade A")
    elif score >= 80:
        print(score, "- Grade B")
    elif score >= 70:
        print(score, "- Grade C")
    else:
        print(score, "- Grade F")
```

```python
books = ["Python Basics", "Data Science"]

title = input("Enter the book title: ")

found = False

for book in books:
    if book == title:
        found = True
        break

if found:
    print("Available")
else:
    print("Not Available")
```

```python
temperatures = [30, 32, 28, 35, 29]

highest = temperatures[0]
lowest = temperatures[0]

for temp in temperatures:
    if temp > highest:
        highest = temp
    if temp < lowest:
        lowest = temp

print("Highest temperature:", highest)
print("Lowest temperature:", lowest)
```

```python
stock = {"Apples": 5, "Bananas": 20}

for item, quantity in stock.items():
    if quantity < 10:
        print(item, "- Low stock (", quantity, ")")
```

```python
seats = [1, 2, 3, 4]
booked_seats = [2, 4]

for seat in seats:
    if seat in booked_seats:
        print("Seat", seat, "→ Booked")
    else:
        print("Seat", seat, "→ Available")
```

```python
balances = [5000, 5200, 5300]
rate = 0.05 / 100   

total_interest = 0

for balance in balances:
    daily_interest = balance * rate
    total_interest += daily_interest
    print("Balance:", balance, "Daily Interest:", daily_interest)

print("Total Interest for the week:", total_interest)
```
